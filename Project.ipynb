```python
from sklearn.model_selection import train_test_split
```

```python
dataNumeric['Purchase'].max()
```

```python
x = dataNumeric.iloc[:,2:11]
```

```python
x
```

```python
y = dataNumeric.iloc[:, 11]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=101)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_pred = regr.predict(x_test)
```

```python
print(y_pred)
```

```python
print('Coefficients: \n', regr.coef_)
```

```python
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
```

```python
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
plt.scatter(y_pred, y_test,  color='black')

plt.plot(y_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

## Alternate way

```python
model = linear_model.LinearRegression(fit_intercept=True, normalize=True, copy_X=True)
```

```python
m = model.fit(x, y)
```

```python
predictions_train = m.predict(x)
print('Score: %.2f' % m.score(x, y))
```

```python
rms_train = sqrt(mean_squared_error(y, predictions_train))
print ('Training set RMSE: %.2f' % rms_train)
```
